## Performed functional, regression, integration, end-to-end, system, and ad-hoc testing.
* **Functional Testing:** - Ensures each feature works according to requirements.
	* **Example**: Login should open the dashboard after entering valid credentials.
	* **Test**: Enter username + password → click login → should open dashboard

* **Regression Testing**: - Existing features still work after a change.
	* **Example**: - After Fixing any bug or after getting new build if any new issue found that was not in old build.

* **Integration Testing:** - how multiple components/modules work together.
	* **Example**: - Thin client + VDI Broker

* **End-To-End Testing:** - Testing the complete workflow from start to finish like a real user.
	* **Example**: - Boot device → login → connect to VDI → open app → logout VDI.

* **Ad-hoc testing:** - Informal testing without test cases.
	* **Example**: - Randomly clicking buttons, entering unexpected values.

* **Sanity Testing:** - After small changes or bug fixes.
	* **Example**: - Bug fixed in login -> only test login.

* **Performance Testing:** - Testing how the system behaves under load, speed, and stress
	* **Example**: -  10 users logging in at same time and Check response time.

* **Smoke Testing:** - Basic Testing to check the build is stable to enough for further testing.
	* **Example**: - App Opens, Login Works.

## Automating test cases where applicable using Python, Selenium, BDD, and tools.
* What is a BDD:
	* It is a Behavior-Driven Development
	* We write clear examples of how the feature should work using Gherkin, a simple language anyone can read.

*	Gherkin Format:
* Feature: User Login
	* Scenario: Successful login with valid credentials
		* Given the user is on the login page
		* When the user enters valid credentials
		* Then the user should be redirected to the dashboard


* Feature: High-level description of a software feature.
* Scenario: A specific example of a behavior (a test case).
* Given: Describes the initial context or pre-condition.
* When: Describes the action or event.
* Then: Describes the expected outcome or result.
* And / But: Adds additional conditions to any of the above steps



## Understanding business requirements and translate them into test scenarios and test cases (EPICS).
* EPIC is a large business requirement or feature that cannot be completed in one go. It is break down into smaller user stories and tasks.
* EPIC - Virtual Desktop Access.
	* User stories: -
		* User should be able to login in
		* User should see available desktops
		* User should launch a session
		* User should disconnect/Reconnect.

* What is Test Scenarios and Test Case?
	* Test scenarios: -
		* Verify user can log in
		* Verify VDI server is reachable
		* Verify Session launch
		* Verify Session reconnect
		* Verify logout

		* Test Case: -
			* Testcase name : - Verify login
			* Steps:
				* Enter username
				* Enter password
				* Click on login button
			* Expected result: - login successful.

* How to translate  EPIC into test scenarios:
	* Understand business requirements
	* Identify scenarios
	* Convert to testcases.
