# Basic Python

### **If-elif-else statement**

```python
a = 1
b = 1
if a > b:
  print("A is greater then B", a)
elif a == b:
  print("Both are equl")
else:
  print("B is greater then A", b)
```

### **Ternary Operators**

```python
a = 2
b = 40
print("A") if a > b else print("B")
```

### **Loops**

```python
i = 0
while i < 10:
  if i == 5:
    break
  print(i)
  i +=1
```

```python
i = 0
while i < 10:
  i += 1
  if i == 4:
    continue
  print(i)
```

```python
list1 = ["Apple", "Banana", "Cherry"]
for i in list1:
  for j in i:
    print(j)
```
