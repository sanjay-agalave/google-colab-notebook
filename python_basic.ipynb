# Basic Python

### **If-elif-else statement**

```python
a = 1
b = 1
if a > b:
  print("A is greater then B", a)
elif a == b:
  print("Both are equl")
else:
  print("B is greater then A", b)
```

### **Ternary Operators**

```python
a = 2
b = 40
print("A") if a > b else print("B")
```

### **Loops**

```python
i = 0
while i < 10:
  if i == 5:
    break
  print(i)
  i +=1
```

```python
i = 0
while i < 10:
  i += 1
  if i == 4:
    continue
  print(i)
```

```python
list1 = ["Apple", "Banana", "Cherry"]
for i in list1:
  for j in i:
    print(j)
```

```python
file = open("Simple.txt", "w")
file.write("Python is Easy and Python is powerful")
file.close()

file = open("Simple.txt", "r")
content = file.read()
content = content.lower()
words = content.split(" ")
print(words)
print(len(words))
word_count = {}

for word in words:
  if word in word_count:
    word_count[word] += 1
  else:
    word_count[word] = 1

for word, count in word_count.items():
  print(word, count)

file.close()
```
