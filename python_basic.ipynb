### **tuples and lists**

```python
class A():
  def add(self, a, b):
    self.a = a
    self.b = b
    return self.a + self.b

class B(A):
  def sub(self, a, b):
    self.a = a
    self.b = b
    return self.a - self.b


obj1 = B()
print(obj1.add(10, 20))
print(obj1.sub(10, 20))
```

```python
class A():
  def add(self, a, b):
    self.a = a
    self.b = b
    return self.a + self.b

class B():
  def sub(self, a, b):
    self.a = a
    self.b = b
    return self.a - self.b

class C(A, B):
  def mul(self, a, b):
    self.a = a
    self.b = b
    return self.a * self.b


obj1 = C()
print(obj1.add(20, 20))
print(obj1.sub(20, 20))
print(obj1.mul(20, 20))
```

```python
class A():
  def add(self, a, b):
    self.a = a
    self.b = b
    return self.a + self.b

class B(A):
  def sub(self, a, b):
    self.a = a
    self.b = b
    return self.a - self.b

class C(B):
  def mul(self, a, b):
    self.a = a
    self.b = b
    return self.a * self.b

obj1 = C()
print(obj1.add(20, 20))
print(obj1.sub(20, 20))
print(obj1.mul(20, 20))
```
