First, we'll import the `sqlite3` module, which is built into Python. Then, we'll connect to a database file. If the file doesn't exist, SQLite will create it.

```python
import sqlite3

# Connect to a SQLite database (creates one if it doesn't exist)
conn = sqlite3.connect('sanjay_database.db')

# Create a cursor object
cursor = conn.cursor()

print("Successfully connected to sanjay_database.db")
```

Now, let's create a table in our new database. We'll use a `try-except` block to handle cases where the table might already exist.

```python
try:
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS users (
            id INTEGER PRIMARY KEY,
            name TEXT NOT NULL,
            email TEXT UNIQUE NOT NULL
        )
    ''')
    conn.commit()
    print("Table 'users' created successfully (or already existed).")
except sqlite3.Error as e:
    print(f"Error creating table: {e}")
```

Next, let's insert some data into the `users` table.

```python
try:
    cursor.execute("INSERT INTO users (name, email) VALUES (?, ?)", ('Alice', 'alice@example.com'))
    cursor.execute("INSERT INTO users (name, email) VALUES (?, ?)", ('Bob', 'bob@example.com'))
    conn.commit()
    print("Data inserted successfully.")
except sqlite3.IntegrityError as e:
    print(f"Error inserting data (possibly duplicate email): {e}")
except sqlite3.Error as e:
    print(f"Error inserting data: {e}")
```

Finally, we can query the data to see what we've stored and then close the connection.

```python
# Query data
cursor.execute("SELECT * FROM users")
rows = cursor.fetchall()

print("\nUsers in the database:")
for row in rows:
    print(row)

# Close the connection
conn.close()
print("\nDatabase connection closed.")
```
